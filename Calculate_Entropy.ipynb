```python
import math
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np

Ey = 0.0
Pa = 0.0
Pb = 0.0

Matrix = [[0 for x in xrange(101)] for y in xrange(101)]

def frange(start, end, increment):
     seed = start
     while seed < end:
     	yield seed
     	seed += increment

itr_range1 = frange(0.00, 1.01, 0.01)

itr_range = []
for i in itr_range1:
    itr_range.append(i)
    #print (itr_range)

for i in range(0,101,1):
    for j in range(0,101,1):
        
        if (((itr_range[i]+itr_range[j]) < 1.0) and ((itr_range[i]+itr_range[j]) != 0.0)) :
            if itr_range[i] == 0.0:
                Matrix[i][j] = - (itr_range[j] * math.log(itr_range[j],2)) - ((1 - itr_range[j]) * math.log((1 - itr_range[j]),2))
            elif itr_range[j] == 0.0:
                Matrix[i][j] = - (itr_range[i] * math.log(itr_range[i],2)) - ((1 - itr_range[i]) * math.log((1 - itr_range[i]),2))
            else:    
                Matrix[i][j] = - (itr_range[i] * math.log(itr_range[i],2)) - (itr_range[j] * math.log(itr_range[j],2)) - ((1 - itr_range[i] - itr_range[j]) * math.log((1 - itr_range[i] - itr_range[j]),2))
        else:
            Matrix[i][j] = 0.0

fig = plt.figure()
ax = fig.gca(projection='3d')
X = np.arange(0.00, 1.01, 0.01)
Y = np.arange(0.00, 1.01, 0.01)
X, Y = np.meshgrid(X, Y)

surf = ax.plot_surface(X, Y, Matrix, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()  
```
