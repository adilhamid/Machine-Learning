Calculate the Entropy of the System and plot the same


```python
import math
from sys import maxint
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import matplotlib.pyplot as plt
import numpy as np


def calculate_Entropy():
    randomval_range = np.arange(0.0, 1.01, 0.01)
    ValueMatrix = [[0 for x in xrange(101)] for y in xrange(101)]

    for i in range(0,101,1):
        for j in range(0,101,1):
            if (((randomval_range[i]+randomval_range[j]) < 1.0) and ((randomval_range[i]+randomval_range[j]) != 0.0)) :
                if randomval_range[i] == 0.0:
                    ValueMatrix[i][j] = - (randomval_range[j] * math.log(randomval_range[j],2)) - 
                    ((1 - randomval_range[j]) * math.log((1 - randomval_range[j]),2))
                elif randomval_range[j] == 0.0:
                    ValueMatrix[i][j] = - (randomval_range[i] * math.log(randomval_range[i],2)) -
                    ((1 - randomval_range[i]) * math.log((1 - randomval_range[i]),2))
                else:    
                    ValueMatrix[i][j] = - (randomval_range[i] * math.log(randomval_range[i],2)) - 
                    (randomval_range[j] * math.log(randomval_range[j],2)) - 
                    ((1 - randomval_range[i] - randomval_range[j]) * math.log((1 - randomval_range[i] - randomval_range[j]),2))
            else:
                ValueMatrix[i][j] = 0.0
    

    figure = plt.figure()
    ax = figure.gca(projection='3d')
    X = np.arange(0.00, 1.01, 0.01)
    Y = np.arange(0.00, 1.01, 0.01)
    X, Y = np.meshgrid(X, Y)

    surf = ax.plot_surface(X, Y, ValueMatrix, rstride=1, cstride=1, cmap=cm.coolwarm, linewidth=0, antialiased=False)
    figure.colorbar(surf, shrink=0.5, aspect=5)
    ax.set_xlabel('P(alpha)')
    ax.set_ylabel('P(beta)')
    ax.set_zlabel('')
    plt.show()  
    return 

calculate_Entropy()

```
