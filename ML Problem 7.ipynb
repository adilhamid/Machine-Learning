```python
#Problem Number 7 Machine Learning Homework
    

# Code for Problem number 7
import numpy as np
import matplotlib.pyplot as plt
import math

def Get_Error_Val(w):
    return ((w - 2)*(w - 4.5)*(w - 5)*(w - 5.5))

def Get_Derivate_Val(w):
    return (4 * pow(w, 3) - 51 * pow(w, 2) + 209.5*w - 273.25)

def plotFuntion(weights):
    t1 = np.arange(1.9, 6.0, 0.01)
    print 'Weigts for plotting', weights
    t2 = np.array(weights)
    
    plt.figure(1)
    plt.subplot(111)
    plt.xlabel('w')
    plt.ylabel('E(w)')
    plt.plot(t1, Get_Error_Val(t1),t2, Get_Error_Val(t2),'ro' )

    plt.show()

def Get_Gradient_Descent(learning_rate, initial_weights):
    counter =0;
    
    for weight in initial_weights:
        counter =0;
        presentVal = Get_Error_Val(weight);
        newVal = 0.0;
        flagInf = False
        newWeight = weight
        
        print 'For Initial Weight: ', weight
        weightList = []
        weightList.append(weight)
        
        while True:
            if(counter>0):
                presentVal = newVal
            newWeight =newWeight - learning_rate * Get_Derivate_Val(newWeight)
            if (-1.9 <= newWeight <= 8.5 ):
                weightList.append(newWeight)
            newVal = Get_Error_Val(newWeight)
           
            counter = counter + 1
            if math.isinf(presentVal):
                flagInf = True
                break
            if(abs(newVal-presentVal) < 0.001):
                break
            print counter, newWeight , presentVal 
        
        if not flagInf:
            print 'Total Number of Iterations needed to Converge is: ', counter
        else:
            print 'The Convergence is not happening'
        print 
        
        plotFuntion(weightList)
        
        print '--------------------------------------------------------------------------------------------'
        
        print
        

weights = [ 2, 3, 4.5, 4.75, 6, 8]
learningRate = 0.05
Get_Gradient_Descent(learningRate, weights)



        
```
