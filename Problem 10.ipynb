```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import random

def fun(x, y):
  return x**2 + y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(-10.0, 10.0, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array([fun(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('w1')
ax.set_ylabel('w2')
ax.set_zlabel('E(w1,w2)')

plt.show()

plt.figure()
cp = plt.contour(X, Y, Z,10)
plt.clabel(cp, inline=True,fontsize=10)
plt.title('Contour Plot')
plt.xlabel('w1')
plt.ylabel('w2')
plt.show()


```

```python
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import random

def fun(x, y):
  return x**2 + y**2/9

def derfun(x,y):
    return 2*x + 2*y/9

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(-10.0, 10.0, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array([fun(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
zg = np.array([derfun(x,y)for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)
ZG = zg.reshape(Z.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('w1')
ax.set_ylabel('w2')
ax.set_zlabel('E(w1,w2)')

plt.show()


plt.figure()
cp = plt.contour(X, Y, Z,6)
plt.clabel(cp, inline=True,fontsize=10)
plt.title('Contour Plot')
plt.xlabel('w1')
plt.ylabel('w2')
plt.show()
```
